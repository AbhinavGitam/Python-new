Ascending and Desscending

```python
num = [100, 77, 99, 86, 23, 25, 41]
num.sort()
print(num)
num.sort(reverse = True)
print(num)
fruit=['orange' , 'mango' , 'apple' , 'banana' , 'Papaya' , 'orange']
thistuple=('brinjal','tomato','orange','orange')
fruit.extend(thistuple)
print(fruit)
fruit.sort()
print(fruit)
fruit.sort(reverse = True)
print(fruit)
```
