```python
import re
def tm(p):
        a = '^[a-z]+_[a-z]+$'
        if re.search(a,  p):
                return 'Found a match!'
        else:
                return('Not matched!')
print(tm("ak_aa"))
print(tm("abra_bba"))
print(tm("AbhiNba_abhi"))
```
