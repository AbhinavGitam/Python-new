# Dictionary

#### Create and print a dictionary:

```python
abhinav = {
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
print(abhinav)
```

#### Print the "brand" value of the dictionary:

```python
abhinav = {
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
print(abhinav["brand"])
```

#### Duplicates Not Allowed

```python
abhinav = {
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000,
  "year": 2020
}
print(abhinav)

# prints the latest value
```

#### Dictionary Length

```python
abhinav = {
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000,
  "year": 2020
}
print(abhinav)
print(len(abhinav))
```

#### Data Types

```python
abhinav = {
  "brand": "bugatti",
  "electric": False,
  "year": 2000,
  "colors": ["red", "white", "blue"]
}
print(abhinav)
```

#### Type

```python
abhinav = {
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
print(type(abhinav))
```

#### Accessing Items

```python
#Type 1

abhinav = {
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
x = abhinav["model"]
print(x)

#Type 2

abhinav =	{
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
x = abhinav.get("model")
print(x)
```

#### To find the keys

```python
abhinav = {
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
x = abhinav.keys()
print(x)
```

```python
abhinav = {
"brand": "bugatti",
"model": "chiron",
"year": 2000
}
x = abhinav.keys()
print(x)
abhinav["color"] = "white"
print(x)
```

#### To get all the values in a code

```python
abhinav = {
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}

x = abhinav.values()

print(x)
```

#### To Get all the items

```python
abhinav = {
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}

x = abhinav.items()

print(x)
```

#### Checking if something exists or not

```python
abhinav = {
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
if "model" in abhinav:
  print("Yes, 'model' is one of the keys in the abhinav dictionary")
```

#### Using the "Update" Function

```python
abhinav = {
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
abhinav.update({"year": 2020})
print(abhinav)
```

#### Adding items in between

```python
abhinav = {
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
abhinav["color"] = "red"
print(abhinav)

# Using the Update Function

abhinav = {
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
abhinav.update({"color": "red"})
print(abhinav)
```

#### Using the pop function

```python
abhinav =	{
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
abhinav.pop("model")
print(abhinav)
```

#### Remove the last inserted item

```python
abhinav =	{
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
abhinav.popitem()
print(abhinav)
```

#### Using the del keyword

#### Clear Function 

```python
abhinav = {
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
abhinav.clear()
print(abhinav)
```

#### Using Loops in dictionaries

```python
# Keys

abhinav =	{
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
for x in abhinav:
  print(x)
print("")
# Values

abhinav =	{
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
for x in abhinav:
  print(abhinav[x])
```

#### Using Values, Keys, item keywords

```python
#Values Keyword

abhinav =	{
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
for x in abhinav.values():
  print(x)
print("")

#Keys Keyword

abhinav =	{
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
for x in abhinav.keys():
  print(x)
print("")

#Item Keyword

abhinav =	{
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
for x, y in abhinav.items():
  print(x, y)
```

#### Copying the dictionaries

```python
#Using copy() function

abhinav = {
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
dictionary = abhinav.copy()
print(dictionary)

#Using dict() function

abhinav = {
  "brand": "bugatti",
  "model": "chiron",
  "year": 2000
}
dictionary = dict(abhinav)
print(dictionary)
```

#### Nested Dictionaries

```python
family = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}
print(family)
```
